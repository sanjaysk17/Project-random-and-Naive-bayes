```python
from imblearn.over_sampling import SMOTE
```

```python
x=df[['sleep_hours_per_day','attendance_percentage','assignments_completed',"parental_education_Master's"]]
y=df['pass_fail']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
sampling=SMOTE(random_state=42)
x_train,y_train=sampling.fit_resample(x_train,y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_trains=scaler.fit_transform(x_train)
x_tests=scaler.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_trains,y_train)
```

```python
y_pred=model.predict(x_tests)
```

Evaluation Metrics

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,roc_auc_score,roc_curve
forheat=confusion_matrix(y_test,y_pred)
sns.heatmap(forheat,annot=True)
```

```python
print("Accuracy Score:",accuracy_score(y_test,y_pred))
```

```python
print("ROC score",roc_auc_score(y_test,model.predict_proba(x_test)[:,1]))
```

```python
from sklearn.ensemble import RandomForestClassifier
modelr=RandomForestClassifier()
modelr.fit(x_train,y_train)
```

```python
y_pred_r=modelr.predict(x_test)
```

```python
y_pred_r
```

```python
forheat=confusion_matrix(y_test,y_pred_r)
sns.heatmap(forheat,annot=True)
```

```python
print("Accuracy Score:",accuracy_score(y_test,y_pred_r))
```

Here With the Reference of confusion matrix,Random Forest is Best on accuracy
